## Model Fit

A model will now be fit and tested using the SARIMA model. Various values for the SARIMA hyperparameters will be tested in order to select the optimal model. A grid search for the best set of parameters will be performed.

### Training and Testing

An approach using a walk-forward validation will be used to optimize the hyperparameters for the model. To assess the performance of the model, a new test set will be obtained from incoming new market values, which is the definition of model testing on unseen data.

### FUNCTIONS

In order to build a functionality that performs walk-forward validation in order to assess different hyperparameters, several functions will be constructed.

#### Fit & Forecast

The function below will fit a SARIMA model and perform a 1-step forecasting.

```python
def sarima_forecast (history, config):
    order, sorder, trend = config
    # Define model
    model = SARIMAX(history,
                    order=order,
                    seasonal_order=sorder,
                    trend=trend,
                    enforce_stationarity=False,
                    enforce_invertibility=False
                    )
    # Fit model
    print('Fitting model...')
    model_fit = model.fit(disp=False, maxiter=5000)
    # Make a 1-step forecast
    print('Making prediction...')
    yhat = model_fit.predict(len(history), len(history)+1)
    print('Prediction:', yhat[0])
    return yhat[0]
```

#### Train/Test Split

```python
def train_test_split (data, n_test):
    print('train_test_split...')
    return data[:-n_test], data[-n_test:]
```

#### RMSE

In order to select the hyperparameters yielding the best-performing model, we need a measure of model performance. Root of meen squared error (RMSE) will be used.

```python
def measure_rmse (actual, predicted):
    print('Calculating error...')
    return sqrt(mean_squared_error(actual, predicted))
```

#### Walk-Forward Validation

Validation function. The idea is to obtain a number of time points to test on (the last time points). For each time point, the model will be trained on the preceding time points and then the current time point will be predicted. The next iteration will move on to the subsequent time point and so on.

```python
def walk_forward_validation (data, n_test, cfg):
    predictions = list()
    # split dataset
    train, test = train_test_split (data, n_test)
    # seed history with training data
    history = [x for x in train]
    # step over each time-step in the test set
    for i in range (len(test)):
        print(f'Making predictions number {i}')
        # fit model and make forecast for history
        yhat = sarima_forecast(history, cfg)
        # store forecast in list of predictions
        predictions.append(yhat)
        # add actual observation to history for next iteration
        history.append(test[i])
    # estimate prediction error
    error = measure_rmse(test, predictions)
    return error
```

#### Score

The function below will serve as a wrapper intended to capture all warning and potential error messages. It is essentially a wrapper for `walk_forward_validation`.

```python
# score a model, return None if failure
def score_model(data, n_test, cfg, debug=False):
    result = None
    # convert config to a key
    key = str(cfg)
    # show warning and fail on exception if debugging
    if debug:
        result = walk_forward_validation(data, n_test, cfg)
    else:
        # one failure during model validation suggest unstable config
        try:
            # never show warning when grid searching
            result = walk_forward_validation(data, n_test, cfg)
        except Exception as e:
            print (f'An error occured: {e}',)
            result = None
        print('\n------------', (key, result), '\n------------\n\n')
    return (key, result)
```

#### Grid Search

Finally, this function will loop over all hyperparameter values to assess model performance. Note that this function can be adapted to apply multi-processing to speed up the grid search.

```python
def grid_search(data, cfg_list, n_test):
    scores = []
    for cfg in cfg_list:
        print(cfg)
        score = score_model(data, n_test, cfg)
        scores.append(score)
    scores = [score_model(data, n_test, cfg) for cfg in cfg_list]
    # sort configs by error, asc
    scores.sort(key=lambda tup: tup[1])
    return scores
```

#### Define hyperparameters

The function below will define lists of hyperparameters to be used for the grid search. These hyperparameters are alluded to [above](#parameters).

<div class="alert alert-block alert-warning">
<b>Parameters!</b> The various values for the hyperparametes are declared directly into the function before the different configurations are extracted. <u>The function can be enhanced by including parameters where one can specify the vaious parameters as arguments when calling the function.</u>
</div>

```python
# Create SARIMA configurations to be tested
def sarima_configs(seasonal=[5, 20]): # seasonality of week and year
    models = list()
    # define config lists
    p_params = [1, 2] # 1 discovered from PACPlot
    d_params = [0] # 0 from using days
    q_params = [0, 1, 2] # Moving average - nothing or last 2 months
    t_params = ['ct','c', 'n'] # The trend is non of these but try
    P_params = [0, 1, 2, 4] # Seasonal autoregressive order (weekly)
    D_params = [0] # Seasonal differencing
    Q_params = [0, 1] # Seasonal adjustment
    m_params = seasonal # Seasonality
    # Create config instances:
    for p in p_params:
        for d in d_params:
            for q in q_params:
                for t in t_params:
                    for P in P_params:
                        for D in D_params:
                            for Q in Q_params:
                                for m in m_params:
                                    cfg=[(p, d, q), (P, D, Q, m), t]
                                    models.append(cfg)
    return models
```

### Grid Search

The functions will now be used to perform a grid search for the vaious configurations for the time series model hyperparameters.  
  
  2 weeks (**10 time points** or 2 cycles) will be used for as validation points.

<div class="alert alert-block alert-info">
<b>NB!:</b> Note that the optimized model will be tested on completely new incoming data. <i>In this context, there is not allocated test dataset.</i>

```python
# transform data
data = list(df['close'])
# data split
n_test = 10
# Generating hyperparameter combinations
cfg_list = sarima_configs()
print('Total number of configurations to test', len(cfg_list))
# grid search
scores = grid_search(data, cfg_list, n_test)
# List top 3 configurations:
for cfg, error in scores[:3]:
    print (cfg, error)
```
